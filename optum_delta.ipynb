```python
%sql 
drop table optum_catelog.optum_schema.hospital_scdt2;
create table optum_catelog.optum_schema.hospital_scdt2 (
  Hospital_id string, 
  Hospital_name string, 
  city string, 
  state string, 
  country string,
  start_date timestamp,
  end_date timestamp,
  is_active string
);
```

```python
%sql
insert into optum_catelog.optum_schema.hospital_scdt2 
select 
hospital_id, 
hospital_name,
city,
state,
country,
current_timestamp() as start_date,
null as end_date,
'Y' as active 
from optum_catelog.optum_schema.hospital_tb;
```

```python
%sql
select * FROM optum_catelog.optum_schema.hospital_tb AS source
LEFT JOIN optum_catelog.optum_schema.hospital_scdt2 AS target
ON source.hospital_id = target.hospital_id 
AND target.is_active = 'Y'  -- Only check active records
WHERE target.hospital_id IS NULL;  
```

```python
%sql  --logic for scd2
MERGE INTO optum_catelog.optum_schema.hospital_scdt2 AS target
USING (
    SELECT hospital_id, hospital_name, city, state, country
    FROM optum_catelog.optum_schema.hospital_tb
) AS source
ON target.hospital_id = source.hospital_id 
AND target.is_active = 'Y' 

WHEN MATCHED AND (target.hospital_name <> source.hospital_name) THEN 
    -- Deactivate old record and set end date
    UPDATE SET target.end_date = current_date(), target.is_active = 'N';

-- Step 2: Insert the new version of changed records
INSERT INTO optum_catelog.optum_schema.hospital_scdt2 (hospital_id, hospital_name, city, state, country, start_date, end_date, is_active)
SELECT source.hospital_id, source.hospital_name, source.city, source.state, source.country, 
       current_date(), NULL, 'Y'
FROM optum_catelog.optum_schema.hospital_tb AS source
LEFT JOIN optum_catelog.optum_schema.hospital_scdt2 AS target
ON source.hospital_id = target.hospital_id 
AND target.is_active = 'Y'  -- Only check active records
WHERE target.hospital_id IS NULL;  -- Ensure new record is inserted
```
